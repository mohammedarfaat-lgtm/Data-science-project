```python

from datetime import datetime

# Example datetime objects
datetime1 = datetime(2024, 12, 23, 14, 30)  # Dec 23, 2024, 2:30 PM
datetime2 = datetime(2024, 12, 23, 18, 15)  # Dec 23, 2024, 6:15 PM

# Calculate the difference
time_difference = datetime2 - datetime1

# Extract the total seconds and convert to hours and minutes
total_seconds = time_difference.total_seconds()
hours = int(total_seconds // 3600)  # Total hours
minutes = int((total_seconds % 3600) // 60)  # Remaining minutes

# Print the result
print(f"Difference: {hours} hours and {minutes} minutes")
```
