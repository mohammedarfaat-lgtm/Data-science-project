```python
# Initialize a dictionary
my_dict = {'shape': 'circle', 'size': 'large'}

# Use setdefault() to set a default value for the key 'color'
default_value = my_dict.setdefault('color', 'red')

# Print the updated dictionary
print("Updated dictionary:", my_dict)

# Print the default value that was set
print("Default value for 'color':", default_value)
```
