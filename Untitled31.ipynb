```python
from datetime import datetime
import pytz

tz = pytz.timezone('US/Eastern')
dt_before = tz.localize(datetime(2024, 3, 10, 1, 30))
dt_after = tz.localize(datetime(2024, 3, 10, 3, 30))
print("Before transition:", dt_before, "After transition:", dt_after)
```
