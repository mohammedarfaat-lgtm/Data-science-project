```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, ne_chunk

```

```python
nltk.download('punkt_tab')       # For tokenization
nltk.download('stopwords')    # For stop words
nltk.download('averaged_perceptron_tagger_eng')  # For part-of-speech tagging
nltk.download('maxent_ne_chunker_tab')  # For named entity recognition
nltk.download('words')       # For word data required by the chunker
nltk.download('wordnet')     # For lemmatization
```

```python
sentence = "Natural Language Processing is a complex field."

```

```python
tokens = word_tokenize(sentence)
print("Tokenized:", tokens)
```

```python
stop_words=set(stopwords.words("english"))
filtered_sentence=[w for w in tokens if not w.lower() in stop_words]
print("stopword removed:",filtered_sentence)
```

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(w) for w in filtered_sentence]
print("Stemmed Words:", stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_sentence]
print("Lemmatized Words:", lemmatized_words)
```

```python


#Tag each word with its part of speech:
pos_tags = pos_tag(filtered_sentence)
print("Part-of-Speech Tags:", pos_tags)
```

```python
ne_chunks = ne_chunk(pos_tags)
print("Named Entities:", ne_chunks)
```
