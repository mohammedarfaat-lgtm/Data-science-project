```python
data=[1,2,2,3,3,3,4,4,4,4]
mean_value=np.mean(data)
mean_value
```

```python
data=[1,2,2,3,3,3,4,4,4,4]
std_dev=np.std(data)
std_dev
```

```python
data=[1,2,2,3,3,3,4,4,4,4]
mode_value=np.mode(data)
std_dev
```

```python
from statistics import mode
data=[1,2,2,3,3,3,4,4,4,4]
mode_value=mode(data)
mode_value
```

```python
skewness=(mean_value-mode_value)/std_dev
skewness
```
