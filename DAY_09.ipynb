```python
# Function Returning Function**: Write a function called `create_adder` that takes a number `x` as a parameter and returns a function that takes another number `y` as a parameter and returns the sum of `x` and `y`.
```

```python
def create_adder(x):
    # The returned function takes y as a parameter and adds x to y
    def adder(y):
        return x + y
    return adder

# Example usage:
add_5 = create_adder(5)  # This creates a function that adds 5 to any number
print(add_5(3))  # Output: 8 (5 + 3)

add_10 = create_adder(10)  # This creates a function that adds 10 to any number
print(add_10(7))  # Output: 17 (10 + 7)
```
