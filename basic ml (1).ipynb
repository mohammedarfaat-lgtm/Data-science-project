```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

```python
sizes=np.array([800,1000,1200,1400,1600]).reshape(-1,1)
prices=np.array([15000,20000,25000,28000,30000])
```

```python
sizes
```

```python
prices
```

```python
model=LinearRegression()
```

```python
model
```

```python
model.fit(sizes,prices)
```

```python
prediction_value=np.array([[10000]])
```

```python
prediction_price=model.predict(prediction_value)
```

```python
prediction_price
```
