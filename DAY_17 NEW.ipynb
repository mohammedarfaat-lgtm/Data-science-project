```python
# Create a pie chart to visualize the number of lok sabha  seats won by each party
```

```python
seats_by_party=dfc.groupby('Ruling Party')['Lok Sabha Seats'].sum().reset_index()
seats_by_party=seats_by_party.sort_values(by='Lok Sabha Seats',ascending=False)

plt.figure(figsize=(10,10))
plt.pie(seats_by_party['Lok Sabha Seats'],labels=seats_by_party['Ruling Party'],autopct="%1.1f%%")
plt.title('Distribution of Lok Sabha Seats By Rulling Party')
plt.axis('equal')
plt.show()
```

```python
# Create a bar chart to visualize the distribution of states based on their Youth percentage with different colors for states ruled by different parties
```

```python
import matplotlib.pyplot as plt

# Define colors for each ruling party
colors = {
    'Bharatiya Janata Party (BJP)': 'Orange',
    'Indian National Congress (INC)': 'green'
}

plt.figure(figsize=(12, 8))
bars = plt.bar(df['State'], df['Youth %'], color=df['Ruling Party'].map(colors).fillna('gray'))
plt.xlabel('State')
plt.ylabel('Youth Percentage')
plt.title('Youth Percentage of Each State (Color-coded by Ruling Party)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Summary Statistics
summary = df[['Population', 'Youth %', 'Divorce rate', 'Unemployment', 'Lok Sabha Seats']].describe()
print(summary)

# Distribution Plots
plt.figure(figsize=(12, 10))
plt.subplot(3, 2, 1)
plt.hist(df['Population'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Frequency')

plt.subplot(3, 2, 2)
plt.hist(df['Youth %'], bins=10, color='salmon', edgecolor='black')
plt.xlabel('Youth %')
plt.ylabel('Frequency')

plt.subplot(3, 2, 3)
plt.hist(df['Divorce rate'], bins=10, color='gold', edgecolor='black')
plt.xlabel('Divorce rate')
plt.ylabel('Frequency')

plt.subplot(3, 2, 4)
plt.hist(df['Unemployment'], bins=10, color='greenyellow', edgecolor='black')
plt.xlabel('Unemployment')
plt.ylabel('Frequency')

plt.subplot(3, 2, 5)
plt.hist(df['Lok Sabha Seats'], bins=10, color='lightcoral', edgecolor='black')
plt.xlabel('Lok Sabha Seats')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
