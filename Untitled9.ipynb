```python
# prompt:    - Remove leading and trailing whitespaces from the string "  hello  ".
# **Trimming Whitespace**:
#    - Remove leading and trailing whitespaces from the string "  hello  ".

string = "  hello  "

# Using strip() method
result = string.strip()
print(result)

# Using lstrip() and rstrip() methods
result = string.lstrip()
result = result.rstrip()
print(result)
```
