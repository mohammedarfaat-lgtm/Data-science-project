```python
import numpy as np
#data set
data=[10,20,40,60,50]
#Calculate the standard deviation
std_dev=np.std(data)
print("the standarad deviation is",std_dev)
#Calculate the mean value
mean_value=np.mean(data)
print("the mean value is",mean_value)
#to check wheather it is spread out or not
if std_dev>mean_value:
  print("the data is spread out")
else:
  print("the data is not spread out")
```
