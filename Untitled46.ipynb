```python
##package
import numpy as np
##dataset
data=[10,20,40,60,50]
std_dev=np.std(data)
print("the standard deviation:",std_dev)


# to calculate mean_value
mean_value=np.mean(data)
print("the mean value is:",mean_value)

#to check weathere its sprad out or not
if std_dev>mean_value:
  print("the data is spread out")
else:
  print("the data is not spread out" )
```
