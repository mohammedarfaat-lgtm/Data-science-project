```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import graphviz

# Step 1: Create the dataset
data = {
    'Weather': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Overcast'],
    'Humidity': ['High', 'Normal', 'High', 'High', 'Normal', 'Normal'],
    'Wind': ['Weak', 'Weak', 'Strong', 'Weak', 'Strong', 'Weak'],
    'Play': ['No', 'Yes', 'Yes', 'No', 'No', 'Yes']
}
df = pd.DataFrame(data)

# Step 2: Convert categorical variables into numerical values
df['Weather'] = df['Weather'].map({'Sunny': 0, 'Overcast': 1, 'Rain': 2})
df['Humidity'] = df['Humidity'].map({'High': 0, 'Normal': 1})
df['Wind'] = df['Wind'].map({'Weak': 0, 'Strong': 1})
df['Play'] = df['Play'].map({'No': 0, 'Yes': 1})

# Step 3: Split data into features (X) and target (y)
X = df[['Weather', 'Humidity', 'Wind']]
y = df['Play']

# Step 4: Train the Decision Tree
clf = DecisionTreeClassifier()
clf = clf.fit(X, y)

# Step 5: Visualize the Decision Tree
dot_data = tree.export_graphviz(clf, out_file=None,
                                feature_names=['Weather', 'Humidity', 'Wind'],
                                class_names=['No', 'Yes'],
                                filled=True, rounded=True,
                                special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("IPL_Play_Decision_Tree")
graph



```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Step 1: Create the dataset
data = {
    'TimeOfDay': ['Morning', 'Afternoon', 'Evening', 'Night', 'Morning', 'Evening'],
    'UpcomingExam': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
    'HomeworkDone': ['Yes', 'Yes', 'No', 'No', 'Yes', 'No'],
    'ShouldStudy': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)

# Step 2: Convert categorical variables into numerical values
df['TimeOfDay'] = df['TimeOfDay'].map({'Morning': 0, 'Afternoon': 1, 'Evening': 2, 'Night': 3})
df['UpcomingExam'] = df['UpcomingExam'].map({'No': 0, 'Yes': 1})
df['HomeworkDone'] = df['HomeworkDone'].map({'No': 0, 'Yes': 1})
df['ShouldStudy'] = df['ShouldStudy'].map({'No': 0, 'Yes': 1})

# Step 3: Split data into features (X) and target (y)
X = df[['TimeOfDay', 'UpcomingExam', 'HomeworkDone']]
y = df['ShouldStudy']

# Step 4: Train the Decision Tree
clf = DecisionTreeClassifier()
clf = clf.fit(X, y)

# Step 5: Make predictions on new data
new_data = pd.DataFrame({
    'TimeOfDay': [0, 1, 2],       # Morning, Afternoon, Evening
    'UpcomingExam': [1, 0, 1],    # Yes, No, Yes
    'HomeworkDone': [1, 1, 0]     # Yes, Yes, No
})

# Predict and convert predictions back to Yes/No
new_data['ShouldStudy'] = clf.predict(new_data)
new_data['ShouldStudy'] = new_data['ShouldStudy'].map({0: 'No', 1: 'Yes'})

# Map numerical values back to categorical for readability
new_data['TimeOfDay'] = new_data['TimeOfDay'].map({0: 'Morning', 1: 'Afternoon', 2: 'Evening', 3: 'Night'})
new_data['UpcomingExam'] = new_data['UpcomingExam'].map({0: 'No', 1: 'Yes'})
new_data['HomeworkDone'] = new_data['HomeworkDone'].map({0: 'No', 1: 'Yes'})

# Print the predictions in table format
print(new_data)


```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Step 1: Create the dataset
data = {
    '     AgeGroup': ['Youth', 'Youth',   'Middle-aged',  'Senior', 'Senior', 'Middle-aged'],
           'Income': ['High',   'Medium', 'Medium',       'Low',    'High',   'Low'],
    'PriorPurchase': ['No',     'Yes',     'Yes',          'No',    'Yes',    'No'],
              'Buy': ['No',      'No',      'Yes',         'No',    'Yes',   'Yes']
}
df = pd.DataFrame(data)

# Step 2: Convert categorical variables into numerical values
df['AgeGroup'] = df['AgeGroup'].map({'Youth': 0, 'Middle-aged': 1, 'Senior': 2})
df['Income'] = df['Income'].map({'Low': 0, 'Medium': 1, 'High': 2})
df['PriorPurchase'] = df['PriorPurchase'].map({'No': 0, 'Yes': 1})
df['Buy'] = df['Buy'].map({'No': 0, 'Yes': 1})

# Step 3: Split data into features (X) and target (y)
X = df[['AgeGroup', 'Income', 'PriorPurchase']]
y = df['Buy']

# Step 4: Train the Decision Tree
clf = DecisionTreeClassifier()
clf = clf.fit(X, y)

# Step 5: Make predictions on new data
new_data = pd.DataFrame({
    'AgeGroup': [0, 2, 1],       # Youth, Senior, Middle-aged
    'Income': [2, 0, 1],         # High, Low, Medium
    'PriorPurchase': [1, 0, 1]   # Yes, No, Yes
})

# Predict and convert predictions back to Yes/No
new_data['Buy'] = clf.predict(new_data)
new_data['Buy'] = new_data['Buy'].map({0: 'No', 1: 'Yes'})

# Map numerical values back to categorical for readability
new_data['AgeGroup'] = new_data['AgeGroup'].map({0: 'Youth', 1: 'Middle-aged', 2: 'Senior'})
new_data['Income'] = new_data['Income'].map({0: 'Low', 1: 'Medium', 2: 'High'})
new_data['PriorPurchase'] = new_data['PriorPurchase'].map({0: 'No', 1: 'Yes'})

# Print the predictions in table format
print(new_data)


```

```python
# Decision Tree - Predict Employee Salary

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Load dataset
data = pd.read_csv("employ salary.txt")

# Split features and target
X = data[['YearsExperience']]
y = data['Salary']

# Create and train model

model = DecisionTreeRegressor(random_state=0)
model.fit(X, y)

# Make prediction
prediction = model.predict([[5.5]])
print("Predicted Salary for 5.5 years of experience:", prediction[0])

# Visualization
X_grid = np.arange(min(X.values), max(X.values), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(X_grid, model.predict(X_grid), color='blue', label='Decision Tree Model')
plt.title('Decision Tree Regression')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
